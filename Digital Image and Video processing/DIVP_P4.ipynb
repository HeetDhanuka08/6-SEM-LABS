```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
def gray_histogram (img):
  print('Original Image')
  cv2_imshow(img)
  print('\n')
  plt.figure(figsize=(10,5))
  plt.hist(img.ravel(),256,[0,256])
  plt.xlabel('Pixel Value')
  plt.ylabel('Frequency')
  plt.title(' Gray Scale Histogram')
  plt.show()
```

```python
lena = cv2.imread('/content/lenagray.jpg')
gray_histogram(lena)
```

```python
cameraman = cv2.imread('/content/cameraman.jpg')
gray_histogram(cameraman)
```

```python
def color_image_histogram(img):
    print('Original Image')
    cv2_imshow(img)
    print('\n')
    print('Color Image Histogram')

    # Separating color channels
    b, g, r = cv2.split(img)

    # Plot histograms for each channel
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 3, 1)
    plt.hist(b.ravel(), 256, [0, 256], color='blue')
    plt.title('Blue Channel Histogram')

    plt.subplot(1, 3, 2)
    plt.hist(g.ravel(), 256, [0, 256], color='green')
    plt.title('Green Channel Histogram')

    plt.subplot(1, 3, 3)
    plt.hist(r.ravel(), 256, [0, 256], color='red')
    plt.title('Red Channel Histogram')

    plt.tight_layout()
    plt.show()
```

```python
lena1 = cv2.imread('/content/lena.jpg')
color_image_histogram(lena1)
```

```python
# Contrast Stretching

alpha = 1.5    # Contrast control (1.0-3.0)
print('Original Image')
cv2_imshow(lena)
print('\n')
Contrast_img = cv2.convertScaleAbs(lena, alpha=alpha)
print('Contrasted Strected Image')
cv2_imshow(Contrast_img)
```

```python
# Brightness stretching

beta = 50    # Brightness control (0-100)
print('Original Image')
cv2_imshow(lena)
print('\n')
brightness_img = cv2.convertScaleAbs(lena, beta=beta)
print('Brightness Strected Image')
cv2_imshow(brightness_img)
```

```python
def histogram_equalization(img):
  # Convert the image to grayscale if it's a color image
  if len(img.shape) == 3:  # Check if the image has 3 channels (color)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert to grayscale

  equ = cv2.equalizeHist(img)
  res = np.hstack((img, equ))
  plt.imshow(res,cmap='grey')
  plt.title('Original vs Equalized Image')
  plt.axis('off')
  plt.show()

  print('\n')

  flat1 = img.flatten()
  flat2 = equ.flatten()
  plt.figure(figsize=(10, 5))
  plt.subplot(1, 2, 1)
  plt.hist(flat1, bins=10, color='grey')
  plt.title('Original Histogram')
  plt.xlabel('Pixel Intensity')
  plt.ylabel('Frequency')

  plt.subplot(1, 2, 2)
  plt.hist(flat2, bins=5, color='grey')
  plt.title('Equalized Histogram')
  plt.xlabel('Pixel Intensity')
  plt.ylabel('Frequency')

  plt.tight_layout()
  plt.show()

```

```python
histogram_equalization(lena)
```
