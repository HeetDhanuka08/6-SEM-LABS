Aim :-To study video processing using opencv python

i) To read and play a video

ii) To save a video

iii) To display the properties of a video



```python
# Reading and Displaying the video
import cv2
from google.colab.patches import cv2_imshow

cap = cv2.VideoCapture('/content/output_video.mp4')

frame_count = 0
while(cap.isOpened() and frame_count < 6):
    ret, frame = cap.read()
    if ret == True:
        cv2_imshow(frame)
        frame_count += 1
    else:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# Saving the video
import cv2
from google.colab.patches import cv2_imshow
import os

cap = cv2.VideoCapture('/content/output_video.mp4')

fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(os.path.join('/content/', 'output_video.mp4'), fourcc, fps, (width, height))
```

```python
#displaying  the properties of a video

import cv2
from google.colab.patches import cv2_imshow

cap = cv2.VideoCapture('/content/output_video.mp4')
print("Frame width:", cap.get(cv2.CAP_PROP_FRAME_WIDTH))
print("Frame height:", cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
print("Frame count:", cap.get(cv2.CAP_PROP_FRAME_COUNT))
print("FPS:", cap.get(cv2.CAP_PROP_FPS))
print("Format:", int(cap.get(cv2.CAP_PROP_FORMAT)))
print("Duration" , cap.get(cv2.CAP_PROP_FRAME_COUNT) / cap.get(cv2.CAP_PROP_FPS))

cap.release()
cv2.destroyAllWindows()
```
