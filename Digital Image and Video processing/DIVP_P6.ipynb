Aim :- To study and perform image morphology using following techniques

i) Dilation

ii) Erosion

iii) Opening

iv) Closing

```python
# Perfroming Dilation on H

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/Letter H.jpg')

kernel = np.ones((3, 3), np.uint8)

dilation = cv2.dilate(img, kernel, iterations=25)
print('original image')
cv2_imshow(img)

print('\nDilated image')
cv2_imshow(dilation)
print(dilation.shape)
```

```python
# Perfroming Dilation on O

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/images O.png')

kernel = np.ones((3, 3), np.uint8)

dilation = cv2.dilate(img, kernel, iterations=5)
print('original image')
cv2_imshow(img)
print('\nDilated image')
cv2_imshow(dilation)
print(dilation.shape)
```

```python
# Perfroming Erosion on H

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/Letter H.jpg')

kernel = np.ones((3, 3), np.uint8)
erosion = cv2.erode(img, kernel, iterations=10)

print('original image')
cv2_imshow(img)
print('\nEroded image')
cv2_imshow(erosion)
print(erosion.shape)
```

```python
# Perfroming Erosion on O
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/images O.png')

kernel = np.ones((3, 3), np.uint8)
erosion = cv2.erode(img, kernel, iterations=10)

print('original image')
cv2_imshow(img)
print('\nEroded image')
cv2_imshow(erosion)
print(erosion.shape)
```

```python
# Perfroming Opening on H

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/Letter H.jpg')

kernel = np.ones((3, 3), np.uint8)

opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel, iterations=10)

print('original image')
cv2_imshow(img)

print('\nOpening image')
cv2_imshow(opening)
opening.shape
```

```python
# Perfroming Opening on O
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/images O.png')

kernel = np.ones((3, 3), np.uint8)

opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel, iterations=30)

print('original image')
cv2_imshow(img)

print('\nOpening image')
cv2_imshow(opening)
opening.shape
```

```python
# Perfroming Closing on H

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/Letter H.jpg')


kernel = np.ones((3,3),np.uint8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel , iterations = 40)


print('original image')
cv2_imshow(img)

print('\nClosing image')
cv2_imshow(closing)
closing.shape
```

```python
# Perfroming Closing on O

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/images O.png')


kernel = np.ones((3,3),np.uint8)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel , iterations = 10)


print('original image')
cv2_imshow(img)

print('\nClosing image')
cv2_imshow(closing)
closing.shape
```
