```python
# Low pass filter of 3x3 on lena
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

lena = cv2.imread('/content/lenagray.jpg')

size = 3
kernel = np.ones((size,size),np.float32)/(size*size)
lp_filtered_image = cv2.filter2D(lena,-1,kernel)

# original image
print('original image')
cv2_imshow(lena)

# low pass filtered image
print(f"\nlow pass filtered image ({size} x {size})")
cv2_imshow(lp_filtered_image)
```

```python
# High pass filter of 3x3 on lena
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

lena = cv2.imread('/content/lenagray.jpg')

size = 3
kernel = np.ones((size,size),np.float32)/(size*size)
hp_filtered_image = lena - cv2.filter2D(lena,-1,kernel) + 128

# original image
print('original image')
cv2_imshow(lena)

# high pass filtered image
print(f"\nhigh pass filtered image ({size} x {size})")
cv2_imshow(hp_filtered_image)
```

```python
# Median filter of 3x3 on cameraman
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

camera = cv2.imread('/content/camera.jpg')

median_filtered_image = cv2.medianBlur(camera, 3)

# original image
print('original image')
cv2_imshow(camera)

# median filtered image
print("\nMedian filtered image (3x3)")
cv2_imshow(median_filtered_image)
```
