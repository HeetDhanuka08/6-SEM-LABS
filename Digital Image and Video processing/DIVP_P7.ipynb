```python
# Horizontal Lines detection using Erode
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/lines.png')
kernal = np.ones((1,20), np.uint8)
eroded_image = cv2.erode(img, kernal, iterations=1)

# Original Image
print("Original Image")
cv2_imshow(img)
# Eroded Image
print("\nEroded Image")
cv2_imshow(eroded_image)
```

```python
# Vertical Lines detection using Erode
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/lines.png')
kernal = np.ones((15,1), np.uint8)
eroded_image = cv2.erode(img, kernal, iterations=1)

# Original Image
print("Original Image")
cv2_imshow(img)
# Eroded Image
print("\nEroded Image")
cv2_imshow(eroded_image)
```

```python
# Canny Edge detector  on Map
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/map.jpg')

edges = cv2.Canny(img,400,600)
# Original Image
print("Original Image")
cv2_imshow(img)

# Canned Image
print("\nCanned Image")
cv2_imshow(edges)
```

```python
# Canny Edge detector  on Tiger
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/tiger.jpg')

edges = cv2.Canny(img,200,300)
# Original Image
print("Original Image")
cv2_imshow(img)

# Canned Image
print("\nCanned Image")
cv2_imshow(edges)
```

```python
# Edge detector on Tiger using prewitt
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/tiger.jpg')
kernelx = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
img_prewittx = cv2.filter2D(img, -1, kernelx)
img_prewitty = cv2.filter2D(img, -1, kernely)

prewittimg = img_prewittx + img_prewitty

# Original Image
print("Original Image")
cv2_imshow(img)

# X image
print("X Image")
cv2_imshow(img_prewittx)

# Y image
print("\nY Image")
cv2_imshow(img_prewitty)

# Prewitt Image
print("\nPrewitt Image")
cv2_imshow(prewittimg)
```

```python
# Hough Transformation

import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

img = cv2.imread('/content/parking_lot.jpg')
# Original Image
print("Original Image")
cv2_imshow(img)

gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray,0,255,apertureSize = 3)

lines = cv2.HoughLinesP(edges,1,np.pi/180,30 , maxLineGap = 250)
for line in lines:
    x1, y1 ,x2, y2 = line[0]

cv2.line(img,(x1,y1),(x2,y2),(255,0,200),1)
images=[img,edges]
plt.figure(figsize=(10,8))

# Transformed Img
print("\nTransformed Image")
for i in range(1, 3):
    plt.subplot(1, 2, i)
    plt.title(f'Image {i}\n')
    plt.imshow(images[i-1], cmap='gray')
```
