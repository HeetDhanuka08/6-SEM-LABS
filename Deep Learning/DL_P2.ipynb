```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('/content/iris.csv')
```

```python
df
```

```python
laer = LabelEncoder()
df['variety'] = laer.fit_transform(df['variety'])
print(df)
```

```python
X = df[['petal.length', 'petal.width']]
Y = df['variety']
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
X_train = X_train.astype(np.float64)
X_test = X_test.astype(np.float64)
```

```python
class MPNeuron:
    def __init__(self):
        self.b = None

    def model(self, x):
        x = pd.to_numeric(x, errors='coerce')
        return (np.sum(x) >= self.b)

    def predict(self, X):
        Y = []
        for index, row in X.iterrows():
            result = self.model(row)
            Y.append(result)
        return np.array(Y)

    def fit(self, X, Y):
        accuracy = {}
        for b in range(X.shape[1] + 1):
            self.b = b
            Y_pred = self.predict(X)
            accuracy[b] = accuracy_score(Y_pred, Y)
        best_b = max(accuracy, key=accuracy.get)
        self.b = best_b
        print('Optimal value of b is', best_b)
        print('Highest accuracy on training set is', accuracy[best_b])
```

```python
mp_neuron = MPNeuron()
mp_neuron.fit(X_train, Y_train)
```

```python

Y_test_pred = mp_neuron.predict(X_test)

accuracy = accuracy_score(Y_test, Y_test_pred)
print("Accuracy on test set:", accuracy)
```
