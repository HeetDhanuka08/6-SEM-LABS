AIM : Write a Python program to detect and extract specific entities from text using Name Entity Recognition.
Write a Python code to perform morphological analysis and lemmatization on the text corpora.

```python
import spacy
nlp = spacy.load("en_core_web_sm")

def extract_entities(text):
    doc = nlp(text)
    entities = []
    for ent in doc.ents:
        entities.append((ent.text, ent.label_))
    return entities


#Example
text = "Apple Inc. was founded by Steve Jobs, Steve Wozniak, and Ronald Wayne in Cupertino, California on April 1, 1976."
extracted_entities = extract_entities(text)

extracted_entities
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

def perform_morphological_analysis(text):
    doc = nlp(text)
    analysis_results = []
    for token in doc:
        analysis_results.append({
            "text": token.text,
            "lemma": token.lemma_,
            "pos": token.pos_,
            "tag": token.tag_,
            "dep": token.dep_,
            "shape": token.shape_,
            "is_alpha": token.is_alpha,
            "is_stop": token.is_stop
        })
    return analysis_results

# Example
text = "Apple Inc. was founded by Steve Jobs, Steve Wozniak, and Ronald Wayne in Cupertino, California on April 1, 1976."
analysis_results = perform_morphological_analysis(text)

analysis_results


```
